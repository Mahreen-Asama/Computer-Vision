```python
import numpy as np
from matplotlib import pylab as plt
import cv2
```

```python
img_ball = cv2.imread('football.png')
#plt.imshow(img_ball)

gray_ball = np.float32(cv2.cvtColor(img_ball, cv2.COLOR_BGR2GRAY))
#plt.imshow(gray_ball,cmap="gray")

smoothed_ball = cv2.bilateralFilter(gray_ball, 5, 75, 75)
#plt.imshow(smoothed_ball,cmap="gray")

corners_ball = cv2.cornerHarris(smoothed_ball, 5, 5, 0.04)
#plt.imshow(corners_ball)

corners_ball = cv2.dilate(corners_ball, None)
#plt.imshow(corners_ball)

img_ball[corners_ball > 0.003 * corners_ball.max()] = [0, 255, 0]
cv2.imwrite("football_result.jpg", img_ball)

plt.imshow(img_ball)
```

## Erosion Vs Dilation

```python
coins_img = cv2.imread('coins.png',0)
kernel = np.ones((5,5), np.uint8)
coins_erosion = cv2.erode(coins_img, kernel, iterations=3)
coins_dilation = cv2.dilate(coins_img, kernel, iterations=3)
#plt.imshow(coins_img)
plt.imshow(cv2.hconcat([coins_erosion, coins_dilation]))
```
