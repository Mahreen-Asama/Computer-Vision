```python
import cv2 as cv
import matplotlib.pylab as plt
colored_img=plt.imread('cat.png')
grey_img=g(colored_img)
plt.imshow(grey_img, cmap="gray")
```

```python
def g(colored_img):          #function to convert colored to gray image and return it
    import numpy as np
    x=colored_img.shape[0]
    y=colored_img.shape[1]

    grey_img=np.zeros((x,y))

    for i in range(x):
        for j in range(y):
            grey_img[i,j]=colored_img[i,j].mean()

    return grey_img
```
