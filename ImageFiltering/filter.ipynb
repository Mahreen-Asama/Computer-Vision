## add gaussian noise to image

```python
img = cv.imread('tree.jpg', 0) #read image in grey sacle more


#   Specifying Gaussian mean and std
mean = 0.0   # some constant
std = 300.0    # some constant (standard deviation)


noisy_img = img + np.random.normal(mean, std, img.shape)
noisy_img_clipped = np.clip(noisy_img, 0, 255)  # we might get out of bounds due to noise

cv.imwrite('noisy_book.png', noisy_img_clipped)
noisy_img_clipped=cv.cvtColor(cv.imread('noisy_book.png'),cv.COLOR_BGR2RGB)
plt.imshow(noisy_img_clipped)
```
