```python
import cv2 as cv
import numpy as np
import matplotlib.pylab as plt

img=cv.imread('shapes.png')

plt.title('hekllo')
plt.xticks([10,20,30])
plt.yticks([0,50,100])

img = cv.imread('shapes.png')

plt.imshow(img)

# to convert the image in grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap="gray")

THRESH = 200
thresh1, res_img1 = cv.threshold(img_gray, THRESH, 255, cv.THRESH_BINARY)
thresh2, res_img2 = cv.threshold(img_gray, THRESH, 255, cv.THRESH_BINARY_INV)
thresh3, res_img3 = cv.threshold(img_gray, THRESH, 255, cv.THRESH_TRUNC)
thresh4, res_img4 = cv.threshold(img_gray, THRESH, 255, cv.THRESH_TOZERO)
thresh5, res_img5 = cv.threshold(img_gray, THRESH, 255, cv.THRESH_TOZERO_INV)

plt.figure(figsize=(20,10))
plt.subplot(231)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(res_img1, cmap='gray')
plt.title('THRESH_BINARY')
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(res_img2, cmap='gray')
plt.title('THRESH_BINARY_INV')
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(res_img3, cmap='gray')
plt.title('THRESH_TRUNC')
plt.xticks([])
plt.yticks([])

plt.subplot(235)
plt.imshow(res_img4, cmap='gray')
plt.title('THRESH_TOZERO')
plt.xticks([])
plt.yticks([])

plt.subplot(236)
plt.imshow(res_img5, cmap='gray')
plt.title('THRESH_TOZERO_INV')
plt.xticks([])
plt.yticks([])

plt.savefig("threshold_results2.png")
plt.show()
```
