```python
import cv2 as cv
import matplotlib.pylab as plt
im=plt.imread('cat.png')
plt.imshow(im)
```

```python
import numpy as np
plt.hist(np.ravel(im),bins=20)
plt.xlabel('gray levels')
plt.ylabel('No. of pixels')
# plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 3, 1)
plt.title("Original")
plt.imshow(im)

fig.add_subplot(1, 3, 2)
plt.title("Rotated")
plt.imshow(np.rot90(im))

fig.add_subplot(1, 3, 3)
plt.title("Rotated")
plt.imshow(np.rot90(np.rot90(im)))
```

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.title("Up down flip")
#   Up down flip
plt.imshow(np.flipud(im))

fig.add_subplot(1, 2, 2)
plt.title("Left Right flip")
#   Left Right Flip
plt.imshow(np.fliplr(im))
```

```python

img_R, img_G, img_B = im.copy(), im.copy(), im.copy()

img_R[:, :, (1, 2)] = 0
img_G[:, :, (0, 2)] = 0
img_B[:, :, (0, 1)] = 0

img_rgb = np.concatenate((img_R,img_G,img_B), axis=1)

plt.figure(figsize=(15, 15))
plt.imshow(img_rgb, cmap="gray")

plt.figure(figsize=(15, 15))
plt.subplot(131)
plt.hist(np.ravel(img_R))
plt.subplot(132)
plt.hist(np.ravel(img_G))
plt.subplot(133)
plt.hist(np.ravel(img_B))

plt.show()
```

```python
#COLORING a portion of image
photo_r = im.copy()
photo_g = im.copy()
photo_b = im.copy()

photo_r[100:200, 200:1000 ,0] = 255
photo_g[100:200, 200:1000 ,1] = 255
photo_b[100:200, 200:1000 ,2] = 255

res = np.hstack((photo_r,photo_g,photo_b))
plt.figure(figsize=(20,10))
plt.imshow(res)
```
