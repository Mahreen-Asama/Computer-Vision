```python
img = cv.imread('apple.jpg')
hist1 = cv.calcHist([img],[0],None,[256],[0,256])
hist2 = cv.calcHist([img],[1],None,[256],[0,256])
hist3 = cv.calcHist([img],[2],None,[256],[0,256])

plt.plot(hist1)
plt.plot(hist2)
plt.plot(hist3)

plt.savefig("dim_hist.png")
plt.show()
```

```python
img = cv.imread('apple.jpg',0)
equ = cv.equalizeHist(img)
res = np.hstack((img,equ)) #stacking images side-by-side
cv.imwrite('res.png',equ)
plt.imshow(res, cmap="gray")
```

```python
hist_eq = cv.imread('res.png')
hist1 = cv.calcHist([hist_eq],[0],None,[256],[0,256])
hist2 = cv.calcHist([hist_eq],[1],None,[256],[0,256])
hist3 = cv.calcHist([hist_eq],[2],None,[256],[0,256])

plt.plot(hist1)
plt.plot(hist2)
plt.plot(hist3)
plt.title("Equalized Histogram")
plt.savefig("equalized_hist.png")
plt.show()
```
