```python
import numpy as np
import cv2 as cv
import matplotlib.pylab as plt
```

```python
img=cv.imread("book.png")
img=cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
print(img.shape)
```

```python

img2=np.zeros((int(img.shape[0]/2), int(img.shape[1]/2), img.shape[2]))

print(img2.shape)

img2

print(img.shape[1])
```

```python
# delete alternate rows------------------

i1=0
j1=0
k1=0
for i in range (img2.shape[0]):
    j1=0
    for j in range (img2.shape[1]):
        img2[i,j]=img[i1,j1]
        j1+=2
        if j1==img.shape[1]:
            break
    i1+=2
    if(i1==img.shape[0]):
        break
        
            
plt.imshow(img2.astype('uint8'))
print(img2.size)
#print(img2)

img2=np.clip(img2,0,255)

cv.imwrite("resize1.png",img2)
```

```python
img3=np.zeros((int(img.shape[0]/2), int(img.shape[1]/2), img.shape[2]))

print(img3.shape)

img3
```

```python
# average alternate pixels------------------

i1=0
j1=0
k1=0
for i in range (img3.shape[0]):
    j1=0
    for j in range (img3.shape[1]):
        img2[i,j]=(img[i1,j1]+img[i1,j1+1])/2
        j1+=2
        if j1==img.shape[1]:
            break
    i1+=2
    if(i1==img.shape[0]):
        break
        
            
plt.imshow(img2.astype('uint8'))
print(img2.size)
#print(img2)

#img2=np.clip(img2,0,255)

cv.imwrite("resize2.png",img2)
```
